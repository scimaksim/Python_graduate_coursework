#### Copying vs. referencing

Be careful when modifying the elements of a compound object. Assignment statements do not copy objects; instead, they create _bindings_ between a target (a spot in computer memory) and an object!

````python
x = [1, 2, 3, "Cats Rule!"]
# New value 'y' points to the memory of 'x'
y = x
print(x, y)
````

````python
# Changing the original modifies both variables!
x[3] = "Dogs Rule!"
print(x, y)
````

If we want to create a copy of an object, we have to use a **method** (in this case, the `copy` method). While functions such as `print(x)` includes the function name and then its arguments, a method is called by referencing the object, appending a period, and adding the method name and any arguments (e.g. `x.copy()`). 

````python
# Create a new memory allocation
y = x.copy()
x[2] = 10
x[3] = "No, cats rule!"
print(x, y)
````

Objects in Python are mutable or immutable (cannot be modified). A list is mutable, a tuple is immutable. 

We cannot change the elements of a tuple unless the element itself is mutable (e.g. if one of the element of a tuple is a list).

````python
# Create a 'tuple' of four values, the last of which is a mutable list
x = (1, 2, 3, ["Cats Rule!"])
y = x
print(x, y)
````

````python
# Modify the object with index 3 (a list)
x[3][0] = "Dogs!"
print(x, y)
````

### If/then/else

**Control flow** refers to the other in which a language executes code. In Python, we have

- conditional logic (`if`, `elif`, and `else`)
- looping (`for`, `while`)

Syntax for `if` looks like

```
if boolean:
    # If boolean is true, execute the chunk of code that is idented
    # Four spaces are recommended but, technically, any indentation can be used
    statement1
    statement2

# Code that is not indented would then execute as normal
```

````python
temp = 30
if temp < 50:
    print(temp, "degrees is cold.")
    print("Wear a jacket outside!")
````

````python
temp = 100
if temp < 50:
    print(temp, "degrees is cold.")
    print("Wear a jacket outside!")
if temp >= 50:
    print(temp, "degrees is not cold.")
````

We also have an `else` block.

```
if boolean:
    execute this code
else: 
    execute this code
```

````python
temp = 100
if temp < 50:
    print(temp, "degrees is cold.")
    print("Wear a jacket outside!")
else:
    print(temp, "degrees is not cold.")
````

We may have multiple blocks to check.

```
if boolean1:
    # if boolean1 is True
    execute this code
elif boolean2:
    # if boolean1 is False, check if boolean2 is True
    # if True
    execute this code
elif boolean3:
    # if boolean1 and boolean2 2 are False, check if boolean3 is True
    # if True
    execute this code
else:
    # if no condition is met
    execute this code
```

````python
temp = 60
if temp < 50:
    print(temp, "degrees is cold.")
    print("Wear a jacket outside!")
elif temp <70:
    print(temp, "degrees is kind of cold...")
    print("You may want to bring an umbrella in case it rains!")
else:
    print(temp, "degrees is not cold.")
````

Lists and tuples are similar. Like strings, lists and tuples are **sequences**, which means they have a logical order to them (first, second, third object, etc.)

Rather than write out the loop, we can use **list comprehensions** (a shorthand!). The format is `[expression for member in iterable]`.

````python
animals = ["Dog", "Cat", "Horse", "Frog", "Cow", "Buffalo", "Deer", "Fish", "Bird", "Fox", "Racoon"]
mylist = []

for x in animals:
    mylist.append(x)
````

````python
# The 'for' loop above can be re-written as
mylist = [x for x in animals]
mylist
````

````python
animals = ["Dog", "Cat", "Horse", "Frog", "Cow", "Buffalo", "Deer", "Fish", "Bird", "Fox", "Racoon"]
mylist = [x for x in animals if "o" in x]
mylist
````

### Dictionaries

Dictionaries have keys and values, or **key-value pairs**. There is no ordering to their elements; instead, we just reference keys.

### Contingency tables with pandas

To make sense of data, we want to understand the types of data and their distributions. 

When we talk about types of data, we have categorical (qualitative) and numerical (quantitative) variables. 

- categorical variables have entries that are a label or an attribute and include subscales such as 
  * nominal
  * ordinal
  
- numerical variables have entries that are values _and_ on which math can be performed (e.g. no zip codes), and include subscales such as
  * discrete
  * continuous





#### Categorical data

The goal with categorical data is to describe the _distribution_ of a variable. A **distribution** is the pattern and the frequency with which we observe that variable (where do we observe it, and how often do we observe it?). 

With categorical variables, we usually want to describe the relative frequency (or count) for each category. We do this using a contingency table, which are the main summary for categorical variables. 

In `pandas`, we can create a one-way contingency table using `.value_counts()` and create multi-way contingency tables using `crosstab()`.

````python
# Consider data on Titanic passengers
import pandas as pd
titanic_data = pd.read_csv("https://www4.stat.ncsu.edu/~online/datasets/titanic.csv")
titanic_data.head()
````

We can then change our three category labels ("C", "Q", "S") into something more descriptive.

##### One-way contingency tables (looking at a single variable by itself)

We now have 3 categorical variables that we are working with - `embarked`, `sex`, and `survived`. How do we summarize them?

We want to summarize their individual distributions. This means we want to talk about _counts_ - how many times did each category occur in the data set? The easiest way to do this is with the `value_counts()` method. 

````python
# Summarize distribution (i.e. get counts)
titanic_data.embarked.value_counts()
````

````python
titanic_data.survived.value_counts()
````

````python
# Here, we add an additional argument - dropna=False.
# pandas uses NaN to represent missing data. By default, with counts,
# pandas drops thosr NaN. 
# This doesn't summarize how many people have missing data,
# which could be important information to know (esp. with many missing observations)
titanic_data.sex.value_counts(dropna=False)
````

##### Two-way contingency tables (learning about the relationship between two variables)

For two-way contingency tables, we switch to the `crosstab()` function. 

````python
# Two-way contingency table.
# Row variable comes first, followd by column variable.
pd.crosstab(titanic_data.embarked, titanic_data.survived)
````

We may want to add **marginal totals**, or **margins**, to these tables. These provide us with summary totals for each variable.  

````python
# Include margins with totals for rows and columns
pd.crosstab(titanic_data.embarked, titanic_data.survived, margins=True)
````

We can also specify our own row names and columns names, rather than using the defaults from the data set.

````python
# Add custom row and column names to two-way contingency table.
# We provide these as a list because sometimes we have 
# more than one row variable and more than one column variable
pd.crosstab(titanic_data.embarked, 
            titanic_data.survived, 
            margins=True,
            rownames=["Embarked Port"],
            colnames=["Survival Status"]
           )
````

##### Three-way contingency table for summarizing three categorical variables

For three-way contingency tables, we can provide multiple rows or multiple columns as a list.

#### Bar plots (categorical data)
We can visualize a contingency table using a _bar plot_. Bar plots take the number of counts from our contingency table and create bars to represent them.

Plotting in python can be overwhelming because of so many choices. The traditional way, however, uses `matplotlib`, which is based on MATLAB plotting. This is similar to base R plotting and requires us to build the plot and lay out everything. Other options for visualization include

We're focusing on categorical variables, which are entries that are either a label or an attribute. Barplts give a visual of their counts. If we're working with `pandas`, we may use the `.plot.bar()` data frame method or `plt.bar()` from `matplotlib`'s pyplot submodule.

````python
# Focus on plotting categorical data

# Import pyplot submodule
import matplotlib.pyplot as plt
import pandas as pd

titanic_data = pd.read_csv("https://www4.stat.ncsu.edu/~online/datasets/titanic.csv")

titanic_data.embarked = titanic_data.embarked.astype("category")
titanic_data.embarked.cat.categories = ["Cherbourg", "Queenstown", "Southampton"]

titanic_data.sex = titanic_data.sex.astype("category")
titanic_data.sex.cat.categories = ["Female", "Male"]

titanic_data.survived = titanic_data.survived.astype("category")
titanic_data.survived.cat.categories = ["Died", "Survived"]
````

- Determine _axes_ and _artist_ elements

Produce the plot and then close it:

- `plt.show()`, then `plt.close()`

With `plt.bar()`, `x` represents the _categories_ and `height` represents the corresponding heights.

##### Stacked bar plot with `matplotlib`

We can create a stacked bar plot, where the colors of the bar are colored by another variable. This allows us to see the relationship between 2 categorical variables at once. 

````python
# Color the bars by another categorical variable

# Create a two-way table
# pd.crosstab() returns a dataframe
stack_table = pd.crosstab(titanic_data.embarked, titanic_data.survived)
stack_table
````

````python
print(titanic_data.embarked.cat.categories)
````

````python
# : means "give me all of my rows", and then "Give my the Survived column"
print(stack_table.loc[:, "Survived"])
````

````python
# Build the plot
plt.subplots(figsize = (12,7))

# Create 2 calls to plt.bar() to draw 2 sets of bars

# Here, x-axis should be the categories, and
# y-axis should be the "Died" one-way table
p1 = plt.bar(
  x = titanic_data.embarked.cat.categories,
  height = stack_table.loc[:, "Died"]
  )

# Here, x-axis should be the categories, and
# y-axis should be the "Survived" one-way table.
# The bottom (where they start) shouldn't be at 0, but should start at 
# the values for the "Died" variable
p2 = plt.bar(
  x = titanic_data.embarked.cat.categories,
  height = stack_table.loc[:, "Survived"],
  bottom = stack_table.loc[:, "Died"]
  )

plt.xlabel("Port Embarked")
plt.ylabel("Number of People")
plt.title("Most Embarked in the Southampton Port \n A higher proportion survived from Cherbourg")
# Need to reference colors in some way, so we called our graphs p1 and p2
plt.legend((p1[0], p2[0]), ('Died', 'Survived'), loc = 0)
plt.show()
plt.close()
````

##### Side-by-side bar plots

We're going to write our own correlation function (Pearson's correlation),

<img src="images/pearsons_corr.png" alt="Drawing" style="width: 500px;"/>

where $(x_i, y_i)$ are numeric variables observed on the same $n$ units, $i = 1, ..., n$.

From our user, we will expect two vectors (function inputs):
- $x$ vector (1D array from `numpy` arryax or `pandas` series)
- $y$ vector (1D array from `numpy` arrays or `pandas` series)

In the function body, we will

- find sample means for $x$ and $y$
- compute numerator sum
- computer denominator sums
- find quotient and return that value

````python
import numpy as np
import pandas as pd

def find_corr(x, y):
    """
    Compute Pearson's Correlation Coefficient
    x and y are assumed to be numpy arrays (1D) or pandas series
    Data is assumed to have no missing values
    """
    xbar = x.mean()
    ybar = y.mean()
    num_sum = sum((x-xbar)*(y-ybar))
    den_sumx = sum((x-xbar)**2)
    den_sumy = sum((y-ybar)**2)
    return num_sum/np.sqrt(den_sumx*den_sumy)
print(find_corr.__doc__)
````

Pearson's correlation can be affected by outliers. Let's "pump up" our function with optional abilities. There is another version of correlaion called "Spearman's correlation". It's similar to Pearson's but, instead of using actual $x$ and $y$ values themselves, it uses the **rankings** of the $x$ and $y$ values. R ranking of 1 is assigned to the smallest value, of 2 to the next smallest, and so on.

In effect, Spearman's gets rid of distances between points and makes them more equal, making he correlation more robust. 

<img src="images/ranked_vars.png" alt="Drawing" style="width: 300px;"/>


The formula for Spearman's is identical to Pearson's, except that the $x$s and $y$s have been replaced with $rx$s and $ry$s.

<img src="images/spearmans_corr.png" alt="Drawing" style="width: 600px;"/>


##### Helper functions

To make the body of our main function clearer and easier to read, we write _helper_ functions. For Spearman's correlation, we will write a helper function to find the ranks of $x$ and the ranks of $y$. 

````python
# Given a 1D array or a pandas series, we will wrtie a helper function to find the ranks.
# We will then call that function within our main function
x = np.array([10, 12, 15, 5, 9, 14])
sortx = sorted(x)
sortx
````

````python
rx = np.ones(len(x)) # default vector of [1, 1, 1, 1, 1, 1]
v = 1 # v will count the ranks

for i in range(0, len(x)):     # loop over the sorted x values
    for j in range(0, len(x)): # loop over the x values
        if sortx[i] == x[j]:   # see if jth x is the sorted x we are on
            rx[j] = v          # assign the appropriate rank in rx[j]
            v += 1             # add one to the rank counter
            break              # no need to keep looking for our sorted x
rx
````

````python
# x is a 1D numpy array or a pandas series
def find_ranks(x):
    """
    Find relative ranks, assumes no ties
    """
    length = len(x)
    sortx = sorted(x)
    rx = np.ones(length)
    v = 1
    for i in range(0, length):
        for j in range(0, length):
            if sortx[i] == x[j]:
                rx[j] = v
                v += 1
                break
    return rx

print(x, find_ranks(x))
````

````python
y = np.array([20, 11, 14, 22, 13, 23])
print(y, find_ranks(y))
````

How do we put this into our original Pearson's function body?

````python
# Put an additional argument, use default method = "Pearson"
def find_corr(x, y, method = "Pearson"):
    """
    Compute Pearson's or Spearman's Correlation Coefficient
    x and y are assumed to be numpy arrays (1D) or pandas series
    Data is assumed to have no missing values
    """
    if method == "Spearman":
        x = find_ranks(x)
        y = find_ranks(y) 
    xbar = x.mean()
    ybar = y.mean()
    num_sum = sum((x-xbar)*(y-ybar))
    den_sumx = sum((x-xbar)**2)
    den_sumy = sum((y-ybar)**2)
    return num_sum/np.sqrt(den_sumx*den_sumy)

find_corr(u, v, method="Spearman")
````

#### Error control

There are two types of errors we may come across.

- syntax errors - when we type something wrong and it can't be parsed by python
- exceptions - occurs when python can't execute our code (during the execution, something bad happens, such as a TypeError)

To handle exceptions, we can utilize the `try` and `except` keywords.

````python
def print_strings(*x): # should all be strings!
    c = 0
    for i in x:
        try:
            print("The value in position " + str(c) + " is: " + i)
        except TypeError:
            print("Oh no! Not a string")
        c += 1

print_strings("cat", "dog", 1, "bird")
````

We can also have multiple `except` statements and an `else` block.

````python
def print_stuff(**x): # now taking key-value pairs
    print(x)
    try:
        print("The supplied name is " + x["name"])
    except TypeError:
        print("Oh no! Not a string")
    except KeyError:
        print("You didn't supply a name!")
        
print_stuff(name = "Maksim", age = "xx")
````

````python
print_stuff(Name = "Maksim", age = "xx")
````

A `finally` clause can be given to always execute at the end.

````python
def print_stuff(**x): # now taking key-value pairs
    print(x)
    try:
        print("The supplied name is " + x["name"])
    except TypeError:
        print("Oh no! Not a string")
    except KeyError:
        print("You didn't supply a name!")
    else: 
        print("You provided the right type of args!")
    finally:
        print("This code ran no matter what")
        
print_stuff(name = "Maksim", age = "xx")
````

Finally, we can create our own exceptions and check for them.

````python
def my_divide(x, y):
    if y == 0:
        raise Exception("Can't divide by 0...")
    return x/y
my_divide(3, 9)
````

````python
my_divide(3, 0)
````

We can even include `else` blocks in our loops.

````python
for x in ["cat", "dog", "bird", "frog"]:
    if "o" in x:
        print(x + " contains an o")
else:
    print("This runs at the end of the loop!")
````
